# 3)Check whether all the zones generate the same amount of sales. Help the company to carry out their study with the help of data provided

Independent variables are also termed as factor. For this experimental design, there is only 1 factor or independent variable to evaluate[Amount of Sales], and therefore, one-way ANOVA method is suitable for analysis.

Null Hypothesis:All the zones generate the same amount of sales

Alternate Hypothesis:All the zones generate the different amount of sales

Significant value, : 0.05 i.e 5%

```python
from scipy.stats import f_oneway
```

```python
import scipy.stats as stats
```

```python
fvalue, pvalue = stats.f_oneway(data['Zone - A'], data['Zone - B'], data['Zone - C'], data['Zone - D'])
print(fvalue, pvalue)
```

From ANOVA test we got f_value = 5.672056106843581 and p_value = 0.0011827601694503335

Insights: The p_value we obtaine dfrom the ANOVA test is smaller than the significant value. Therefore we reject the null hypothesis.
 
Therefore we can conclude that zones generates different amount of sales.
